## Programs without arguments

```python
def math_operation():
    num1=eval(input("Enter first number:"))
    num2=eval(input("Enter second number:"))
    sub=num1-num2
    mul=num1*num2
    div=num1/num2
    print(f"The differencec between {num1} and {num2} is: {sub}")
    print(f"The product of {num1},{num2} is: {mul}")
    print(f"The division of {num1} and {num2} is: {div}")
math_operation()
```

```python
import math
def radius_of_circle():
    r=eval(input("Enter the radius of the circle:"))
    area=round(math.pi*r*r)
    print(f"The area of circle with radius {r} is: {area}")
radius_of_circle()
```

```python
def area_of_traingle():
    h=eval(input("Enter the height of the triangle:"))
    b=eval(input("Enter the breadth of the triangle:"))
    area_of_traingle=round(0.5*h*b)
    print(f"The area of triangle with height {h} and breadth {b} is: {area_of_traingle}")
area_of_traingle()
```

```python
def average_of_nums():
    n1=eval(input("Enter first number:"))
    n2=eval(input("Enter second number:"))
    n3=eval(input("Enter third number:"))
    average=round((n1+n2+n3)/3)
    print(f"The average of {n1}, {n2} and {n3} is: {average}")
average_of_nums()
```

```python
def tip_calculator():
    bill=eval(input("Enter the bill amount:"))
    tip=eval(input("Enter the tip given:"))
    total_bill=bill+tip
    print(f"The total bill paid is: {total_bill}")
tip_calculator()
```

```python
def tax_calculator():
    sal=eval(input("Enter the salary:"))
    tax_percent=eval(input("Enter the tax percentage:"))
    tax_paid=sal*tax_percent/100
    print(f"The tax amount paid for salary {sal} with tax percentage {tax_percent} is: {tax_paid}")
tax_calculator()
```

```python
def currency_calculator():
    in_dollars=eval(input("Enter the currency in dollars:"))
    in_rupees=in_dollars*80
    print(f"The amount in rupees for dollars {in_dollars} is: {in_rupees}")
currency_calculator()
```

```python
def weight_converter():
    in_kg=eval(input("Enter the weight in kgs:"))
    in_pound=in_kg*2.2
    print(f"The weight in pounds is: {in_pound}")
weight_converter()
```

## Programs with arguments

```python
def math_operations(num1,num2):
    sub=num1-num2
    mul=num1*num2
    div=num1/num2
    print(f"The differesnce between {num1} and {num2} is: {sub}")
    print(f"The product of {num1} and {num2} is: {mul}")
    print(f"The division of {num1} and {num2} is: {round(div)}")
math_operations(10,12)
```

```python
import math
def radius_of_circle(r):
    area=round(math.pi*r*r)
    print(f"The area of circle with radius {r} is: {area}")
radius_of_circle(100)
```

```python
def area_of_traingle(h,b):
    area_of_traingle=round(0.5*h*b)
    print(f"The area of triangle with height {h} and breadth {b} is: {area_of_traingle}")
area_of_traingle(15,30)
```

```python
def average_of_nums(n1,n2,n3):
    average=round((n1+n2+n3)/3)
    print(f"The average of {n1}, {n2} and {n3} is: {average}")
average_of_nums(10,12,14)
```

```python
def tip_calculator(bill,tip):
    total_bill=bill+tip
    print(f"The total bill paid is: {total_bill}")
tip_calculator(1000,500)
```

```python
def tax_calculator(sal,tax_percent):
    tax_paid=sal*tax_percent/100
    print(f"The tax amount paid for salary {sal} with tax percentage {tax_percent} is: {tax_paid}")
tax_calculator(10000,10)
```

```python
def currency_calculator(in_dollars):
    in_rupees=in_dollars*80
    print(f"The amount in rupees for dollars {in_dollars} is: {in_rupees}")
currency_calculator(100)
```

```python
def weight_converter(in_kg):
    in_pound=in_kg*2.2
    print(f"The weight in pounds is: {in_pound}")
weight_converter(70)
```
