```python
height=eval(input("Enter the height in cms:"))
if height>0:
    lenght_in_inches=round(height*2.54)
    print(f"The height {height} cms in inches is:{lenght_in_inches}")
else:
    print("Enter a valid height")
```

```python
temp=eval(input("Enter the temperature:"))
units=input("What units is the input temperature?:").lower()
if units=="celsius":
    temp_in_fahrenheit=round((temp*9/5)+32)
    print(f"The temperature {temp} in fahreinheit is:{temp_in_fahrenheit}")
else:
    temp_in_celcius=round((temp-32)*5/9)
    print(f"The temperature {temp} in celcius is:{temp_in_celcius}")
```

```python
temp=eval(input("Enter the tempersture in celsius:"))
if temp<-273.15:
    print(f"The temperature {temp} is invalid because it is below zero")
elif temp==-273.15:
    print(f"The temperature {temp} is absolutely zero")
elif temp>-273.15 and temp<0:
    print(f"The temperature {temp} is below freezing")
elif temp==0:
    print(f"The temperature {temp} is at freezing point")
elif temp>0 and temp<100:
    print(f"The temperature {temp} is normal")
elif temp==100:
    print(f"The temperature {temp} is at boiling point")
else:
    print(f"The temperature {temp} is above boiling point")
```

```python
no_of_credits=eval(input("Enter the no of credits you have taken:"))
if no_of_credits<=23:
    print("You are a freshman.")
elif no_of_credits>23 and no_of_credits<=53:
    print("You are asophomore.")
elif no_of_credits>53 and no_of_credits<=83:
    print("You are junior.")
else:
    print("You are senior.")
```

```python
year=eval(input("Enter a year:"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(f"The year {year} is a leap year.")
else:
    print(f"The year {year} is not a leap year.")
```
