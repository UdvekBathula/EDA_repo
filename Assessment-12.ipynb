```python
def num_sum(a,b,c):
    sum=a+b+c
    print(f"The sum of {a},{b},{c} is: {sum}")
num_sum(10,20,30)
```

```python
def radius_of_circle():
    r=eval(input("Enter the radius of the circle:"))
    pi=3.14
    area=round(pi*r*r)
    print(f"The area of circle with radius {r} is: {area}")
radius_of_circle()
```

```python
def even_odd(num):
    if num%2==0:
        return(f"{num} is even.")
    else:
        return(f"{num} is odd.")
even_odd(10)
```

```python
def bmi_calculator(weight,height):
    bmi=weight/(height**2)
    return round(bmi,2)
bmi_calculator(70,1.75)
```

```python
def greet(name,age):
    return(f"Hello {name},you are {age} years old!")
greet("Omkar",30)
```

```python
def max_of_nums(a,b):
    if a-b>0:
        return a
    else:
        return b
max_of_nums(4,9)
```

```python
def power(base,exponent):
    result=base**exponent
    return result
power(2,3)
```

```python
def simple_intrest(principal,rate,time):
    interest=(principal*rate*time)/100
    return interest
simple_intrest(1000,5,2)
```
